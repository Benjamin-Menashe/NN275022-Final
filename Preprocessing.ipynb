```python
import torch
from torch import nn
import numpy as np

import pandas as pd
import matplotlib as plt
from google.colab import drive

drive.mount('/content/drive')
np.random.seed(314)
```

```python
df = pd.read_csv("drive/MyDrive/Colab_Notebooks/NN_project/Clean_4.csv")
```

```python
sub_list = df['Participant_ID'].unique()
x = []

for i in sub_list:
  for t in range(1,56):
   #cur = df.query(f"Participant_ID == '{i}' & Text_ID=={t}").filter(items=['Word_Length','Syl','Content','LSA_Context_Score','Noun','Verb','Adjective','Preposition','Article','Adverb','Pronoun','Determiner','Conjunction','LSA_Context_Score','Freq_log','Orth_log','Orth_F_log','IA_FIXATION'])
   cur = df.query(f"Participant_ID == '{i}' & Text_ID=={t}").filter(items=['Word_Length','Syl','Content','LSA_Context_Score','Noun','Verb','Adjective','Preposition','Article','Adverb','Pronoun','Determiner','Conjunction','LSA_Context_Score','Freq_log','Orth_log','Orth_F_log','N1_F_log','N1_C_log','N2_F_log','N2_C_log','N3_F_log','N3_C_log','UN1_F_log','UN1_C_log','UN2_F_log','UN2_C_log','UN3_F_log','UN3_C_log','IA_FIXATION'])
   x += [torch.tensor(cur.values)]
```

```python
xp = nn.utils.rnn.pad_sequence(x, batch_first=True, padding_value=0.0)
xp = torch.nan_to_num(xp)
```

```python
train_size = int(0.7 * len(xp))
test_size = len(xp) - train_size
train_dataset, test_dataset = torch.utils.data.random_split(xp, [train_size, test_size])
```

```python
test_size = int(0.5 * test_size)
valid_size = len(test_dataset) - test_size
test_dataset, valid_dataset = torch.utils.data.random_split(test_dataset, [test_size, valid_size])
```

```python
train_loader = torch.utils.data.DataLoader(dataset=train_dataset, 
                                           batch_size=64, 
                                           shuffle=True)

valid_loader = torch.utils.data.DataLoader(dataset=valid_dataset, 
                                           batch_size=64, 
                                           shuffle=True)

test_loader = torch.utils.data.DataLoader(dataset=test_dataset, 
                                           batch_size=64, 
                                           shuffle=True)
```

```python
torch.save(train_loader, 'drive/MyDrive/Colab_Notebooks/NN_project/train_loader.pth')
torch.save(valid_loader, 'drive/MyDrive/Colab_Notebooks/NN_project/valid_loader.pth')
torch.save(test_loader, 'drive/MyDrive/Colab_Notebooks/NN_project/test_loader.pth')
```
